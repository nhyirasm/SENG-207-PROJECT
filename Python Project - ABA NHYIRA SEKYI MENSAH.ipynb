```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### THE DATASET

```python
# Load dataset
df = pd.read_csv('data.csv')
df
```

### Check for null values

```python
# Check for null values
print(df.isnull().sum())
```

### Remove null values


```python
# Remove null values
df.dropna(inplace=True)
```

##### Null values of this dataset were removed because the number of null values as compared to the entire dataset were relatively small hence, removing these null values would not have much of an effect on the dataset. Also, the mean of columns with null values was not used as some columns may have outliers in their dataset and might not give a very accurate analysis of the data

### Check for duplicates

```python
# Check for duplicates
print('Number of duplicates:', df.duplicated().sum())
```

```python
# Histogram of alcohol column
sns.histplot(df['alcohol'], kde=False, bins=20, color='blue')
plt.title('Histogram of Alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()

```

```python
# Histogram of pH column
sns.histplot(df['pH'], kde=False, bins=20, color='green')
plt.title('Histogram of pH')
plt.xlabel('pH')
plt.ylabel('Frequency')
plt.show()

```

```python
# Histogram of Fixed Acidity column
sns.histplot(df['fixed acidity'], kde=False, bins=20, color='purple')
plt.title('Histogram of Fixed Acidity')
plt.xlabel('fixed acidity')
plt.ylabel('Frequency')
plt.show()

```

```python
# Histogram of quality column
sns.histplot(df['quality'], kde=False, bins=20, color='orange')
plt.title('Histogram of Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.show()

```

```python
# Histogram of pH column
sns.histplot(df['residual sugar'], kde=False, bins=20, color='red')
plt.title('Histogram of residual sugar')
plt.xlabel('residual sugar')
plt.ylabel('Frequency')
plt.show()

```

```python
# Histogram of pH column
sns.histplot(df['density'], kde=False, bins=20, color='brown')
plt.title('Histogram of density')
plt.xlabel('density')
plt.ylabel('Frequency')
plt.show()

```

```python
# Scatter plot of alcohol vs pH
sns.scatterplot(x='alcohol', y='pH', data=df)
plt.title('Scatter Plot of Alcohol vs pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()

```

```python
# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
